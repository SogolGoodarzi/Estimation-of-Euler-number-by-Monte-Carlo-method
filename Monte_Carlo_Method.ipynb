# Part 3: Estimation of 'e' by Monte_Carlo methode

# part C:

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random,exp
n=100
def Estimator(n):
    x,y=[],[]
    x=random.uniform(0,1,size=n)
    y=random.uniform(np.exp(-1),1,size=n)
    count=0
    for i in range(0,n):
        if(y[i]<np.exp((-1)*x[i])):
            count+=1
    return(x,y,count/n)
def plot(x,y):
    plt.scatter(x,y,color="green")
    for i in range(0,len(x)):
        if(y[i]<np.exp((-1)*x[i])):
            plt.scatter(x[i],y[i],color="red")
    t=np.arange(0,1,0.0001)
    plt.plot(t,np.exp((-1)*t),color="black")
[x,y,est]=Estimator(n)
print("Probability of red dots:",est)
plot(x,y)
e=1/(1-est)+1
print("Value of estimation for e:",e)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random,exp
n=100
def Estimator(n):
    x,y=[],[]
    x=random.uniform(0,1,size=n)
    y=random.uniform(np.exp(-1),1,size=n)
    count=0
    for i in range(0,n):
        if(y[i]<np.exp((-1)*x[i])):
            count+=1
    return(count/n)
Est,Sigma=[],[]
for i in range(0,n):
    e=1/(1-Estimator(n))+1
    Est.append(e)
print("Expectation:",(sum(Est)/n))
for i in range(0,n):
    Sigma.append((Est[i]-sum(Est)/n)**2)
print("Variance:",(sum(Sigma)/n))
```

# part E:

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random,exp
n=100
N=np.arange(100,1001,50)
def Estimator(n):
    x,y=[],[]
    x=random.uniform(0,1,size=n)
    y=random.uniform(np.exp(-1),1,size=n)
    count=0
    for i in range(0,n):
        if(y[i]<np.exp((-1)*x[i])):
            count+=1
    return(count/n)
Exp,Est=[],[]
for n_ in N:
    e=[]
    for i in range(0,n):
        e.append(1/(1-Estimator(n_))+1)
    Est.append(e)
    Exp.append(sum(e)/n)
print("Expectation:",Exp)
Variance=[]
for n_ in range(0,len(N)):
    var=[]
    for i in range(0,n):
        var.append((Est[n_][i]-Exp[n_])**2)
    Variance.append(1/n*sum(var))
print("Variance:",Variance)
z=[]  #list for confidence interval
for j in range(0,len(N)):
    z.append(1.96*np.sqrt(Variance[j]/n))
plt.plot(N,Exp)
for i in range(0,len(N)):
    plt.plot([N[i],N[i]],[Exp[i]-z[i],Exp[i]+z[i]])
plt.xlabel("n")
plt.ylabel("Expectation and 95% confidence interval")
plt.figure()
yerr=z
plt.errorbar(N, Exp, yerr,fmt='o',color='purple')
plt.plot(N,Exp,color="orange")
plt.xlabel("n")
plt.ylabel("Expectation and 95% confidence interval")
```
